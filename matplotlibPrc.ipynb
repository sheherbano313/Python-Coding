```python
#lab 24 Bar chart
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style  
subjects=['English','AI','DIP','PL','IS']
sect_A=[80,90,87,23,34]
sect_B=[20,30,17,33,12]
plt.bar(subjects,sect_A)
```

```python
plt.barh(subjects,sect_B)
```

```python
plt.bar(subjects,sect_A,width=0.4,color='g',align='edge',edgecolor='r',linewidth=5,label='sect_A')
plt.legend()
```

```python
style.use('ggplot')
plt.bar(subjects,sect_A,width=0.4,color='g',align='edge',edgecolor='r',linewidth=5,label='sect_A')
plt.legend()
plt.title('sect_A',fontsize=20)
plt.xlabel('Subjects')
plt.ylabel('Students')
```

```python
style.use('ggplot')
plt.bar(subjects,sect_A,width=0.4,color='g',alpha=0.4,label='sect_A')
plt.bar(subjects,sect_B,width=0.4,color='r',alpha=0.6,label='sect_A')
plt.legend()
plt.title('sect_A & B Details',fontsize=20)
plt.xlabel('Subjects')
plt.ylabel('Students')
```

```python
style.use('ggplot')
sub_index=np.arange(len(subjects))
width=0.4
plt.bar(sub_index,sect_A,width,color='g',alpha=0.4,label='sect_A')
plt.bar(sub_index+width,sect_B,width=0.4,color='r',alpha=0.6,label='sect_A')
plt.legend()
plt.title('sect_A & B Details',fontsize=20)
plt.xlabel('Subjects')
plt.ylabel('Students')
```

```python
style.use('ggplot')
sect_C=[5,10,15,20,25]
sub_index=np.arange(len(subjects))
width=0.4
plt.bar(sub_index,sect_A,width=0.4,color='g',alpha=0.4,label='sect_A',align='edge')
plt.bar(sub_index+width,sect_B,width=0.4,color='r',alpha=0.6,label='sect_B',align='center')
plt.bar(sub_index,sect_C,width=0.4,color='b',alpha=0.2,label='sect_C',align='edge')
plt.legend()
plt.title('sect_A ,B & C Details',fontsize=20)
plt.xlabel('Subjects')
plt.ylabel('Students')
```
